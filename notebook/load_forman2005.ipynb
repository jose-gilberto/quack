```python
import re
import pandas as pd

arff_file = "./forman03a_data/fbis.mat.arff"
data_started = False
attributes = []
rows = []

with open(arff_file) as f:
    for line in f:
        line = line.strip()
        if line.startswith("@ATTRIBUTE"):
            # Pega apenas o nome do atributo
            attributes.append(line.split()[1])
        elif line.upper().startswith("@DATA"):
            data_started = True
        elif data_started and line:
            # Inicializa com zeros
            row = [0] * len(attributes)
            
            # Exemplo: {0 1,5 1}
            matches = re.findall(r"(\d+)\s+([\d.]+)", line)
            for idx, val in matches:
                row[int(idx)] = float(val)
            
            rows.append(row)

df = pd.DataFrame(rows, columns=attributes)
```
